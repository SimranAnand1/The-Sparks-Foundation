```python
# Importing the required libraries 
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
pipeline=kr.pipeline.Pipeline()
```

```python
# Extracting the images
image_ocr=[ kr.tools.read(i) for i in ['img1.jpg', 'img2.jpg','img3.jpg','img4.jpg','img5.jpg']]
```

```python
# Number of images
len(image_ocr)
```

```python
# Plotting the first image
plt.figure(figsize=(5,5))
plt.imshow(image_ocr[0])
```

```python
# Plotting the second image
plt.figure(figsize=(5,5))
plt.imshow(image_ocr[1])
```

```python
# Plotting the third image
plt.figure(figsize=(5,5))
plt.imshow(image_ocr[2])
```

```python
# Plotting the fourth image
plt.figure(figsize=(5,5))
plt.imshow(image_ocr[3])
```

```python
# Plotting the fifth image
plt.figure(figsize=(5,5))
plt.imshow(image_ocr[4])
```

```python
prediction_groups=pipeline.recognize(image_ocr)
```

###Predictions

```python
# Implementing the character detector
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(20,20))
for ax, image, predictions in zip(axs, image_ocr, prediction_groups):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
